```python
# Create a list of printable ASCII characters from 32 to 126
visible_ascii = [chr(i) for i in range(32, 127)]  # 95 characters

# Extended printable ASCII characters (from 160 to 255, excluding non-printable ones)
extended_ascii = [
    chr(i) for i in range(161, 323) if i not in {173}  # Exclude non-visible characters like 173, 180
]

# Combine the lists to ensure we have exactly 256 characters
all_visible_characters = visible_ascii + extended_ascii

# Ensure the total length is 256
print("Number of characters:", len(all_visible_characters))

# Convert the characters to their ASCII values
ascii_values = [ord(char) for char in all_visible_characters]

# Display the list of ASCII values
print("Character ASCII values:", ascii_values)

#for val in ascii_values:
#    print (chr(val), val)
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt



def flatten(img):
    #arr = [img.shape[0]*img.shape[1]]
    arr = []
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            arr.append(img[i, j])
    return arr



def img_extractor(path):    

    main_img_arr = []
    for i in range(0,len(ascii_values)):
        p = path + str(ascii_values[i]) + '.png'
        img = cv2.imread(p, 0)
        print(type(img))
        main_img_arr.append(flatten(img))

    return main_img_arr


main_img_arr = img_extractor(os.getcwd()+"/ascii_characters_cal/")



```
